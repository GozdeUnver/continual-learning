```python
import matplotlib.pyplot as plt

# Data from the table
data = {
    "Background": {"Cadis-train": 70.02, "Cadis-val": 71.15, "Cadis-test": 68.34, "Cataract-train": 83.99, "Cataract-val": 84.17, "Cataract-test": 84.01},
    "Knife": {"Cadis-train": 0.24, "Cadis-val": 0.2, "Cadis-test": 0.35, "Cataract-train": 0.03, "Cataract-val": 0.02, "Cataract-test": 0.02},
    "Bonn forceps": {"Cadis-train": 0.25, "Cadis-val": 0.15, "Cadis-test": 0.45, "Cataract-train": 0.02, "Cataract-val": 0.03, "Cataract-test": 0.01},
    "Cannula": {"Cadis-train": 0.28, "Cadis-val": 0.25, "Cadis-test": 0.31, "Cataract-train": 0.15, "Cataract-val": 0.14, "Cataract-test": 0.17},
    "Capsulorhexis Cystotome": {"Cadis-train": 0.09, "Cadis-val": 0.08, "Cadis-test": 0.11, "Cataract-train": 0.04, "Cataract-val": 0.02, "Cataract-test": 0.04},
    "Capsulorhexis forceps": {"Cadis-train": 0.08, "Cadis-val": 0.02, "Cadis-test": 0.07, "Cataract-train": 0.1, "Cataract-val": 0.12, "Cataract-test": 0.1},
    "Phacoemulsification handpiece": {"Cadis-train": 0.17, "Cadis-val": 0.24, "Cadis-test": 0.24, "Cataract-train": 0.75, "Cataract-val": 0.47, "Cataract-test": 0.56},
    "Micromanipulator": {"Cadis-train": 0.19, "Cadis-val": 0.23, "Cadis-test": 0.24, "Cataract-train": 0.35, "Cataract-val": 0.26, "Cataract-test": 0.31},
    "I/A handpiece": {"Cadis-train": 0.31, "Cadis-val": 0.55, "Cadis-test": 0.47, "Cataract-train": 0.68, "Cataract-val": 0.7, "Cataract-test": 0.75},
    "Lens injector": {"Cadis-train": 0.31, "Cadis-val": 0.38, "Cadis-test": 0.36, "Cataract-train": 0.15, "Cataract-val": 0.25, "Cataract-test": 0.2},
    "Pupil": {"Cadis-train": 17.04, "Cadis-val": 15.7, "Cadis-test": 16.38, "Cataract-train": 7.48, "Cataract-val": 7.45, "Cataract-test": 7.21},
    "Iris": {"Cadis-train": 11.02, "Cadis-val": 11.1, "Cadis-test": 12.69, "Cataract-train": 6.25, "Cataract-val": 6.36, "Cataract-test": 6.63}
}

# Function to create pie charts with a legend and remove labels for classes with less than 1%
def create_pie_chart(dataset, title):
    labels = list(data.keys())
    sizes = [data[label][dataset] for label in labels]
    colors = plt.get_cmap('tab20')(range(len(labels)))  # Use a colormap with many distinct colors

    # Filter out labels with less than 1%
    filtered_labels = [label if data[label][dataset] >= 1 else '' for label in labels]

    fig, ax = plt.subplots()
    wedges, texts, autotexts = ax.pie(sizes, colors=colors, labels=filtered_labels, autopct=lambda pct: f'{pct:.1f}%' if pct >= 1 else '', startangle=140)
    ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

    plt.legend(wedges, labels, title="Classes", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    # plt.title(title)
    plt.savefig(f"imgs/{dataset.lower()}-dist.png", bbox_inches='tight')
    plt.show()

# Create pie charts for each dataset
create_pie_chart("Cadis-train", "Class Distribution in CaDIS Train Set")
create_pie_chart("Cadis-val", "Class Distribution in CaDIS Validation Set")
create_pie_chart("Cadis-test", "Class Distribution in CaDIS Test Set")
create_pie_chart("Cataract-train", "Class Distribution in Cataract-1K Train Set")
create_pie_chart("Cataract-val", "Class Distribution in Cataract-1K Validation Set")
create_pie_chart("Cataract-test", "Class Distribution in Cataract-1K Test Set")
```
